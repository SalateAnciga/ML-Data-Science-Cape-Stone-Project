```python
dataset=pd.read_csv("teen_phone_addiction_dataset.csv")
```

```python
#Distribution plot-UNIVARIATE
```

```python
sb.distplot(dataset['Academic_Performance'],hist=True,kde=True,kde_kws={'color':'blue'})
```

```python
sb.histplot(dataset['Academic_Performance'],kde=True)
```

```python
sb.distplot(dataset['Academic_Performance'],hist=False,kde=True,rug=True)
```

```python
#Distribution plot-BIVARIATE
```

```python
sb.jointplot(x="Time_on_Social_Media",y="Addiction_Level",data=dataset,kind="hex")
plt.show()
```

```python
sb.jointplot(x="Time_on_Social_Media",y="Addiction_Level",data=dataset,kind="kde")
plt.show()
```

```python
#PAIRPLOT
```

```python
sb.pairplot(dataset,hue='Gender',diag_kind="kde",kind="hist",palette="husl")
plt.show()
```

```python
#For CATEGORICAl DATA
```

```python
sb.stripplot(x="Phone_Usage_Purpose",y="Addiction_Level",data=dataset,hue="Gender")
plt.show()
```

```python
sb.swarmplot(x="Gender",y="Time_on_Social_Media",data=dataset)
plt.show()
```

```python
#Distribution plot-categoricaldata vs numercialdata
```

```python
sb.violinplot(x="Gender",y="Addiction_Level",data=dataset)
plt.show()
```

```python
sb.barplot(x="Gender",y="Time_on_Education",data=dataset)
plt.show()
```

```python
sb.catplot(x="Phone_Usage_Purpose",y="School_Grade",data=dataset,hue="Gender",kind="violin")
plt.show()
```
