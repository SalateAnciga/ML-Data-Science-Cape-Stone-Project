```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
dataset=pd.read_csv("teen_phone_addiction_dataset.csv")
```

```python
dataset
```

```python
stats.f_oneway(dataset['Daily_Usage_Hours'],dataset['Sleep_Hours'],dataset['Weekend_Usage_Hours'])
```

```python
#Reject Null hypothesis p<0.05
#Accept  Alternative hypothsis
#H0:There is no significance differance between these columns-Null Hypothesis
#H1:There is a significance differance between these columns-Alternative hypothesis
```

```python
#pvalue is greater than 0.05 so we accept alternative hypothsis.Moreover there is a significance differance between these columns.
```

```python
two_Way=ols('Addiction_Level ~ Gender + School_Grade + Location + Gender*  School_Grade',data=dataset).fit()
```

```python
Two=sm.stats.anova_lm(two_Way,typ=2)
```
